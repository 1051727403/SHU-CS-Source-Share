# K-means聚类算法

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
from scipy.io import loadmat
```

## 1. 二维数据集的处理

### 1.1 导入数据，并封装成数据结构处理

```python
two_dimension_data = loadmat('data/2D_data.mat')  # two_dimension_data为字典
X = two_dimension_data['X']  # X为2维ndarray对象
data1 = pd.DataFrame(X, columns=['X1', 'X2']) # data1为datafrmae对象
data1 # 查看data1
```

### 1.2 观察数据分布

```python
sb.set(context="notebook", style="darkgrid")
sb.lmplot(x = 'X1', y = 'X2', data=data1, fit_reg=False)
plt.show()
```

### 1.3 手写函数定义K-means（4个函数）

```python
def init_centroids(X, k): # 初始化簇坐标（选择数据集中的样本）
    m, n = X.shape # m = 300 n = 2 m为数据个数，n为数据维度
    centroids = np.zeros((k, n))
    idx = np.random.randint(0, m, k) # k为簇类数
    
    for i in range(k):
        centroids[i,:] = X[idx[i],:]
    
    return centroids # (k,n)的numpy数组
```

```python
def find_closest_centroids(X, centroids):  # 寻找簇坐标
    m = X.shape[0]  # 300
    k = centroids.shape[0]  # 3
    idx = np.zeros(m)  # 为每一个样本点设置空id数组

    for i in range(m):  # 两层循环，寻找坐标
        min_dist = 1000000  # 设置最小距离默认值
        for j in range(k):
            dist = np.sum((X[i, :] - centroids[j, :]) ** 2)  # 计算
            if dist < min_dist:
                min_dist = dist
                idx[i] = j

    return idx  # 返回size为m的一维数组
```

```python
def compute_centroids(X, idx, k): # X为2维数组， idx为 k为簇的数量
    n = X.shape[1]  # n = 2 数据维度
    centroids = np.zeros((k, n)) # 聚类中心坐标
    
    for i in range(k):
        indices = np.where(idx == i) # 返回满足条件的数组索引
        centroids[i,:] = (np.sum(X[indices,:], axis=1) / len(indices[0])).ravel()
    
    return centroids
```

```python
def run_k_means(X, initial_centroids, max_iters):  # 进行迭代 max_iters为最大迭代次数
    m = X.shape[0]
    k = initial_centroids.shape[0]
    idx = np.zeros(m)
    centroids = initial_centroids
    for i in range(max_iters):
        idx = find_closest_centroids(X, centroids)
        centroids = compute_centroids(X, idx, k)

    return idx, centroids
```

### 1.4 测试手写的函数

```python
init = init_centroids(X, 3) # 多次运行，产生坐标随机
init
```

```python
idx = find_closest_centroids(X, init) # 运行后发现对于每一个样本点，都对应1个簇
idx
```

```python
idx, centroids = run_k_means(X, init, 10) # 测试封装好的run_k_means函数
idx, centroids
```

```python
fig, ax = plt.subplots(1, 2, figsize=(20, 8))

plt.sca(ax[0])  # 切换到子图
plt.scatter(x='X1', y='X2', data=data1)
plt.xlabel("Raw Data", fontsize=20)

plt.sca(ax[1])
cluster1 = X[np.where(idx == 0)[0], :]  # 簇1
cluster2 = X[np.where(idx == 1)[0], :]  # 簇2
cluster3 = X[np.where(idx == 2)[0], :]  # 簇3

plt.scatter(cluster1[:, 0], cluster1[:, 1], s=30, color='r', label='Cluster 1')
plt.scatter(cluster2[:, 0], cluster2[:, 1], s=30, color='g', label='Cluster 2')
plt.scatter(cluster3[:, 0], cluster3[:, 1], s=30, color='b', label='Cluster 3')
plt.legend()
plt.xlabel("K-means Clustering", fontsize=20)

plt.show()
```

## 2. 图像的处理(使用sklearn中封装好的函数)

### 2.1 鸟

#### 2.1.1 数据处理

```python
from skimage import io
import matplotlib.pyplot as plt

# 除以浮点数，保证画质衰减没有那么多
pic_bird = io.imread('data/bird_small.png') / 255.
pic_bird
# io.imshow(pic_bird)
# plt.show()
# pic_bird.shape
```

```python
data_bird = pic_bird.reshape(128*128, 3) # 对数据进行序列化
# data_bird.shape
data_bird
```

#### 2.1.2 训练模型

```python
from sklearn.cluster import KMeans#导入kmeans库

model_list = [] # 设置两个列表，分别存储model pic
pic_list = []

model = KMeans(n_clusters=15, n_init=100)
model.fit(data_bird)
centroids = model.cluster_centers_
C = model.predict(data_bird) # 返回属于簇的缩影
compressed_pic = centroids[C].reshape((128,128,3))
pic_list.append(compressed_pic)
model_list.append(model)

model = KMeans(n_clusters=5, n_init=100)
model.fit(data_bird)
centroids = model.cluster_centers_
C = model.predict(data_bird)
compressed_pic = centroids[C].reshape((128,128,3))
pic_list.append(compressed_pic)
model_list.append(model)

model = KMeans(n_clusters=3, n_init=100)
model.fit(data_bird)
centroids = model.cluster_centers_
C = model.predict(data_bird)
compressed_pic = centroids[C].reshape((128,128,3))
pic_list.append(compressed_pic)
model_list.append(model)
```

#### 2.1.3 绘图

```python
fig, ax = plt.subplots(1, 4, figsize = (20 , 8))

ax[0].imshow(pic_bird)
plt.sca(ax[0])
plt.xlabel("Original Image", fontsize = 20)
plt.xticks([])
plt.yticks([])

ax[1].imshow(pic_list[0])
plt.sca(ax[1])
plt.xlabel("Clusters = 15", fontsize = 20)
plt.xticks([])
plt.yticks([])


ax[2].imshow(pic_list[1])
plt.sca(ax[2])
plt.xlabel("Clusters = 5", fontsize = 20)
plt.xticks([])
plt.yticks([])

ax[3].imshow(pic_list[2])
plt.sca(ax[3])
plt.xlabel("Clusters = 3", fontsize = 20)
plt.xticks([])
plt.yticks([])

plt.show()
```
